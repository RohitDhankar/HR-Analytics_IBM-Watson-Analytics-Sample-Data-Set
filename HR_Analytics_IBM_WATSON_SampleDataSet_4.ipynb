```python
from sklearn.model_selection import StratifiedShuffleSplit

    
'''
Source :-- http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html#sklearn.model_selection.StratifiedShuffleSplit

1st RUN -- 
__________________________________
StratifiedShuffleSplit(n_splits=3, random_state=123, test_size=0.3,
            train_size=None)
            
            

'''

print "__________________________________"

sss = StratifiedShuffleSplit(n_splits=3, test_size=0.3, random_state=123) # test_size=0.3 thus TRAIN_size =0.7 OR 70% 
sss.get_n_splits(X, y)

print(sss)       

print "__________________________________"

for train_index, test_index in sss.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index) # Printing INDEX Values not ACTUAL Feature Values 
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
print "__________________________________"

print X_train.shape
print y_train.shape
print X_test.shape
print y_test.shape
```
